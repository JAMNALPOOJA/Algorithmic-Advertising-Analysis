# Model Building

```python
# Simpe Linear Regression

X = df['TV']
Y = df['Sales']
validation_size= 0.3
seed= 42
```

```python
# Statsmodels library fits a line on the dataset which passes through the origin
import statsmodels.api as sm
```

```python
#Train_Test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size = validation_size, random_state= seed)
```

```python
lr = LinearRegression()
```

```python
# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the regression Line using 'OLS'(Ordinary Least Squares)
lr = sm.OLS(Y_train, X_train_sm).fit()

# Adding a constant to X_test
X_test_sm = sm.add_constant(X_test)

# Predict the y values corresponding to X_test_sm
Y_pred = lr.predict(X_test_sm)
```

```python
lr.params         # parameters/coefficients , i.e the intercept and the slope of the regression line fitted
```

```python
print(lr.summary())
```

```python
# Inference: R- squared is 0.57 i.e. 57% of the variance in sales is explained by TV.
```

# Linear regression equation:
Sales = 7.239460  + 0.046408 X TV

```python
plt.scatter(X_test, Y_test)
plt.plot(X_test,7.239460 + 0.046408 * X_test, 'r' )
plt.show()
```

# Model Evaluation

```python
import sklearn.metrics as metrics
```

```python
print ('MAE: {}'.format(metrics.mean_absolute_error(Y_test, Y_pred)))
print('MSE:{}'.format(metrics.mean_squared_error(Y_test, Y_pred)))
print('RMSE:{}'.format(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred))))
```
